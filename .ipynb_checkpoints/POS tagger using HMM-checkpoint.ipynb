```python
#Creating two sets unique_words and unique_states to store the unique words and the unique labels respectively
unique_words = set()
unique_states = set()
```

```python
def insert_words(i_list):
    unique_words.update(i_list)
def insert_states(i_list):
    unique_states.update(i_list)
    
```

```python
df["sentence"].apply(insert_words)
df["labels"].apply(insert_states)
unique_words.add("<UNK>")
unique_states = sorted(unique_states)
unique_words = sorted(unique_words)
```

```python
#Calculating the total number of unique words and the total number of unique states
n_words = len(unique_words)
n_states = len(unique_states)
print(n_words)
print(n_states)
```

```python
'''
Creating a trasition matrix here which is basically a matrix of size ((n_states) * n_states) 
where the rows depict the previous state and the columns depict the current state. The cells
depict the probability of a state occuring if the current state if the previous state is the state depicted by the row
Also creating a start_matrix which will store the initial probabilities of all the states
'''
transition_matrix = np.zeros((n_states, n_states))
start_matrix = np.zeros(n_states)
dict_states = {} #this is a dictionory that maps every state to a number
for i, temp in enumerate(unique_states):
    dict_states[temp] = i
    
for j, row in df.iterrows(): #Populating the cells with appropriate count of every state
    label = row["labels"]   
    for i, temp in enumerate(label):
        if(i == 1):
            start_matrix[dict_states[temp]]+=1
        else:
            transition_matrix[dict_states[label[i - 1]]][dict_states[temp]]+=1 
        
            
```

```python
transition_matrix
```

```python
transition_matrix = transition_matrix/transition_matrix.sum(axis=1)[:,None]
transitionView = pd.DataFrame(transition_matrix)
transitionView
```

```python

'''creating an emission matrix here which is basically a matrix of size (n_words * n_states) 
where the rows depict the current word and the columns depict the current state. The cells
depict the probability of a state occuring if the current state if the previous state is the state depicted by the row
'''
emission_matrix = np.zeros((n_words, n_states))
dict_words = {} 
for i, temp in enumerate(unique_words):
    dict_words[temp] = i
    
for j, row in df.iterrows(): #Populating the cells with appropriate count of every state
    label = row["labels"]   
    sentence = row["sentence"]
    for i, temp in enumerate(label):
            emission_matrix[dict_words[sentence[i]]][dict_states[label[i]]]+=1  
```

```python
emission_matrix
```

```python
emission_matrix = emission_matrix/emission_matrix.sum(axis=0)#Dividing all the columns by the sum of the columns
emissionView = pd.DataFrame(emission_matrix)
emissionView
```

```python
start_matrix
```

```python
#Dividing all the rows by the sum of the row
start_matrix = start_matrix/start_matrix.sum()
startView = pd.DataFrame(start_matrix)
startView
```

<p> These three matrices namely, transition probability matrix, emission probability matrix and start probability matrix help in <br> 
     determining the hidden states of a sequence if the observable states are given. The transition probabilty matrix is the probability<br>
     of a state y to occur if the previous state was y0, or transition_matrix[y0][y] = P(t = y|t-1 = y0). The emmision probability matrix<br>
     is the probabilty of an observable state to appear given its hidden state, or exmission_matrix[x][y] = P(x|y). Finally the start <br>
     probabilty matrix represents the probability of a state to appear if it is the first in the sequence or, start_matrix[y] = P(-1|t0 = y) </p>

<p> Once all the matrices have been defined we need to use an appropriate algorithm to make our predictions, for this I chose the Viterbi algorithm.
    So a basic dry run where we have a sequence of 3 words for example, "I am Dev", and we have 4 different tags, noun, preposition, verb and adjective, we will be trying out 4<sup>3</sup>combinations. We will be calculating the probability of each unique combination occuring using the Baye's rule and then selecting the path with the highest probability. The formula turns out to be<br> P(t - 1|y0) * P(x|y) * P(t = y|t-1 = y0)). However this brute force method is very complex and impractical. Here, we can use the Viterbi algorithm which basically uses the dynamic programming by saving states 
and avoiding overlapping subproblems. I have implemented the same below.</p>

```python
def viterbi(obs, states, start_p, trans_p, emit_p):
    '''
    here,   
        obs is the current sequence
        states is a list of unique states
        start_p is the starting probability matrix
        emit_p is the emission probability matrix
    '''
    V = [{}] #This is basically the dp arry which is a list of dictionaries
    path = {}
    #Initialising the base case here, which is when we are at the first word
    for y in states:
        if(obs[0] in dict_words): #this if statement is added to handle the case when a word which was not in our training corpus is encountered
            V[0][y] = start_p[dict_states[y]] * emit_p[dict_words[obs[0]]][dict_states[y]]
            path[y] = [y]
        else:
            V[0][y] = start_p[dict_states[y]] * emit_p[dict_words["<UNK>"]][dict_states[y]] #"<UNK>" is an extra row in the emission matrix 
            path[y] = [y]  

    for t in range(1, len(obs)):#iterating through every word
        V.append({})
        newpath = {}

        for y in states: #iterating through every state here
            temp = emit_p[dict_words["<UNK>"]][dict_states[y]] #again handling the case where the word is not in our training corpus
            if(obs[t] in dict_words):
                temp = emit_p[dict_words[obs[t]]][dict_states[y]]
            
            (prob, state) = max(
                [(V[t-1][y0] * trans_p[dict_states[y0]][dict_states[y]] * temp, y0) for y0 in states]
            ) #calculating the max probability considering every previous state
            V[t][y] = prob
            newpath[y] = path[state] + [y]

        path = newpath 

    (prob, state) = max([(V[-1][y], y) for y in states])
    return (prob, path[state])
```

```python
def laplace_smoothing(transition_matrix, smoothing_factor=1.0):

  num_states = transition_matrix.shape[0]
  smoothed_matrix = transition_matrix.copy()
  smoothed_matrix += smoothing_factor / (num_states * (num_states - 1))
  smoothed_matrix /= (1 + smoothing_factor)
  return smoothed_matrix

smoothed_transition_matrix = laplace_smoothing(transition_matrix)


```

```python
def em_smoothing(matrix, smoothing_factor=1.0, vocabulary_size=None):
  
    vocabulary_size = matrix.shape[1]  # Infer from matrix dimensions
    num_states = matrix.shape[0]  # Number of rows (states)
    smoothed_matrix = matrix.copy()
    smoothed_matrix += smoothing_factor / (num_states * vocabulary_size)
    smoothed_matrix /= (1 + smoothing_factor)
    return smoothed_matrix

```

```python
smoothed_emission_matrix = em_smoothing(emission_matrix)
smoothed_start_matrix = laplace_smoothing(start_matrix)
```

```python
correct_predictions = 0 #declaring a variable to count the correct predictions 
for i,obs in enumerate(test_obs):
    pred = viterbi(obs,unique_states,start_matrix,smoothed_transition_matrix,smoothed_emission_matrix)# I have used these matrices after applying smoothing measures on them
    predictions = pred[1] #extracting the path from the tuple which is the predicted value
    actual = test_actual[i] #this is the actual value
    for j,temp in enumerate(predictions):
        if(temp == actual[j]):
            correct_predictions+=1; #incrementing the correct predictions every time a correct tag is predicted in the sequence 
        
        
```
